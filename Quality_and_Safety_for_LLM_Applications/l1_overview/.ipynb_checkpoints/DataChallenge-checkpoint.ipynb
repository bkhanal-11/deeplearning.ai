## Summarizing our data

Fifty data points is very reasonable to work with, but what if we're doing this at scale? For a production application, there may be millions of data points each day. So how do we summarize this data?

There are three basic options when datasets are too large to reasonably store and analyze: **simple data profiling**, **more advanced data profiling**, and **random sampling**.

First up, random sampling:

```python
sample = chats.sample(frac=0.1)
sample
```

Next, data profiling:

```python
profile = chats.describe()
profile
```

Finally, more advanced sketch-based data profiling techniques. And for that, we'll look at whylogs:

```python
import whylogs as why
```

```python
why.init(session_type="whylabs_anonymous")
```

```python
whyprofile = why.log(chats, name="original data, default whylogs").profile()
```

```python
whyprofile.view().to_pandas().T
```

But there's more metrics that you can collect. Specifically for text and LLMs, we've released the open-source LangKit with many relevant metrics for text data:

```python
from langkit import llm_metrics
```

```python
schema = llm_metrics.init()
```

```python
langprofile = why.log(chats, name="original data, textstat", schema=schema).profile()
```

Let's click the link to see our results and many additional features from the WhyLabs AI Observability platform.

```python
langprofile.view().to_pandas()
```

Now, let's visualize locally:

```python
from whylogs.viz import NotebookProfileVisualizer
```

```python
visualizer = NotebookProfileVisualizer()
```

```python
visualizer.set_profiles(langprofile.view())
```

```python
visualizer.double_histogram("prompt.character_count")
```
